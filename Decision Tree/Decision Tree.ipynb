```python
import pandas as pd
```

```python
dataset = pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
 # Converting true or False to 0/1
dataset = pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset = dataset.astype(int)
```

```python
dataset
```

```python
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
```

```python
dependent = dataset[['Profit']]
```

```python
from sklearn.model_selection import train_test_split #train_test_split is used to train the ML for Linear regression
X_train,X_test,Y_train,Y_test = train_test_split(independent,dependent,test_size =0.30,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion = 'absolute_error',splitter = 'best') #Hyper Tuning Parameters
regressor = regressor.fit(X_train,Y_train)
```

```python
#For Creating graph we are using Matplot Library
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
Y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score #Evaluation Metrics
r_score =r2_score(Y_test,Y_pred)
```

```python
r_score
```

```python
#Finalize the model and save using pickle library
import pickle
filename = "finalized_model_Decision_Tree.sav" #pickle should always be stored in sav format
pickle.dump(regressor,open(filename,'wb')) #wb - Write Binary
```

```python
loaded_model = pickle.load(open("finalized_model_Decision_Tree.sav",'rb'))
result = loaded_model.predict([[1234,456,789,1,0]]) #For this model we need to give 5 inputs
```
